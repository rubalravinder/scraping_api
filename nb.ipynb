## Récup d'images

```python
# Récupérer une liste d'images au format JSON

list_img = requests.get('https://picsum.photos/v2/list')
list_img_json = list_img.json()
list_img_json
```

```python
list_img.headers['Content-Type'] # Content-Type donne le type de données qu'on récupère
```

```python
# Récupérer les 50 premières images de la requête en changeant le param limit dans la requête HTML

list_of_50_fst_imgs = requests.get('https://picsum.photos/v2/list?page=2&limit=50')
```

```python
# Liste des images où le prénom de l’author contient la lettre r (stockées dans un fichier json)

list_r_author_imgs = [img for img in list_of_50_fst_imgs.json() if 'r' in img['author']]
list_r_author_imgs_url = [img['url'] for img in list_r_author_imgs]
```

```python
# Télécharger 10 des images de la liste d'url fraîchement créée

for img in list_r_author_imgs[:10]:
    r = requests.get(img['download_url']) # on stocke la requête
    # r.content -> imbuvable à lire car c'est les pixels de l'image en binaire
    with open((f"downloaded_imgs/{img['id']}.jpg"), "wb") as file: # on ouvre un fichier où on écrira en binaire
        file.write(r.content)
        file.close()

# Les images sont stockées sur le pc
```
